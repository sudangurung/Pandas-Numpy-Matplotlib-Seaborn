```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data.
url = "https://raw.githubusercontent.com/Pitsillides91/Python-Tutorials/master/Seaborn%20Tutorial/Marketing%20Raw%20Data.csv"
data = pd.read_csv(url, sep=",")

# Show the first 5 rows.
data.head(5)
```

```python
# Simple Line Graph. 
# Assume we want to investigate the revenue by date. 

ax = sns.lineplot(x="Week_ID",y="Revenue",data=data)
# Need to specify our x and y axis and the data frame we want to use.
# Note: Error bands show the confidence interval. 
```

```python
# Example 2: Adding Categories (by promo). 
ax = sns.lineplot(x="Week_ID", y="Revenue", data=data, hue="Promo")
```

```python
# Example 3: Promo with style. 
ax = sns.lineplot(x="Week_ID",y="Revenue",data=data, hue="Promo", style="Promo")
```

```python
# Example 4: By Promo with style & increase the size & remove the CI. 

#increase the size. 
sns.set(rc={"figure.figsize":(12,10)})
ax = sns.lineplot(x="Week_ID",y="Revenue",data=data, hue="Promo", style="Promo", ci=False)
```

```python
# Example 5: By promo with style & increase the size & Remove CI & adding markers & By Month

ax = sns.lineplot(x="Month_ID",y="Revenue",data=data, hue="Promo", style="Promo", ci=False, markers=True)
```

```python
# Example 6: By Day_name with style & increase the size & Remove CI & adding markers & By Month

ax = sns.lineplot(x="Month_ID",y="Revenue",data=data, hue="Day_Name", style="Promo", ci=False, markers=True)
```

```python
# Example 6: By Year with style & increase the size & Remove CI & adding markers & By Month

ax = sns.lineplot(x="Year",y="Revenue",data=data, hue="Day_Name", style="Promo", ci=False, markers=True)
```

```python
# Bar Plots
# Example 1: Total Revenue by Month

ax = sns.barplot(x="Month_ID",y="Revenue",data=data)
#Note: The bar plots takes mean as default. 

# Create a df with the mean values. 
data[["Revenue","Month_ID"]].groupby("Month_ID").agg({"Revenue":"mean"})
```

```python
# Example 2: Total Revenue by month & Remove the confidence interval. 
ax = sns.barplot(x="Month_ID",y="Revenue",data=data, ci=False)
```

```python
# Example 3: Total Revenue by month & Remove the CI & By promo. 
ax = sns.barplot(x="Month_ID",y="Revenue",data=data, ci=False, hue="Promo")
```

```python
# Example 5: Total Revenue by month & Remove the CI & by Promo & change the direction (orient). 
ax = sns.barplot(x="Revenue",y="Month_ID",data=data,ci=False,hue="Promo",orient="h")
```

```python
# Example 6: Total Revenue by month & Remove the CI & by Promo & change the direction (orient) & change colour. 
ax = sns.barplot(x="Revenue",y="Month_ID",data=data,ci=False,hue="Promo",orient="h",color="blue")
```

```python
# Histogram
# Example 1: Investigate the distribution of Revenue. 
x = data["Revenue"].values

sns.distplot(x, color="blue")
```

```python
#Example 2: Investigate the distribution of Revenue, adding the mean
mean = data["Revenue"].mean()
mean

sns.distplot(x, color="blue")

plt.axvline(mean,0,1, color="red")
```

```python
# Example 3: Investigate the distribution of visiors, adding the mean
x = data["Visitors"].values

mean = data["Visitors"].mean()

sns.distplot(x, color="red")

plt.axvline(mean,0,1, color="black")

plt.xlabel("Revenue")
plt.ylabel("Frequency")
plt.title("Revenue Distribution")
```

```python
# Box Plot
# Example 1: Investigate the distribution of Revenue. 
x = data["Revenue"].values

ax = sns.boxplot(x, color="blue")

#Notes: 
# The line signifies the median
#The box in the middle show the beginning of Q1 (25th percentile) and the end of Q3 (75th percentile)
# The whiskers (left - right) show the minimum quartile and maximum quartile
# The dots on the right are "outliers"
```

```python
# Example 2: Investigate the distribution of revenue by day.
ax = sns.boxplot(x="Day_Name", y="Revenue", data=data)
```

```python
# Example 3: Investigate the distribution of revenue by day - horizontal - change colors.
ax = sns.boxplot(x="Revenue", y="Day_Name", data=data, color="blue")
```

```python
# Example 4: Investigate the distribution of revenue by day - change color - add hue (promo)
ax = sns.boxplot(x="Day_Name", y="Revenue", data=data, color="red", hue="Promo")
```

```python
# Example 4: Investigate the distribution of revenue by day - change color - add data points(swarmplot)
ax = sns.boxplot(x="Day_Name", y="Revenue", data=data, color="yellow")
ax = sns.swarmplot(x="Day_Name", y="Revenue", data=data, color="black")
```

```python
#lM Plots
# Example 1: Relationship between marketing spend and revenue. 
ax = sns.lmplot(x="Marketing Spend", y="Revenue", data=data, height=10)
```

```python
#Example 2: Relationship between marketing spend and Revenue - changing color, hue & style
ax = sns.lmplot(x="Marketing Spend", y="Revenue", data=data, height=10, hue="Promo", ci=False, markers=["+","o","x"])
```

```python
# Example 3: Relationship between Marketing Spend and Revenue - by column.
ax = sns.lmplot(x="Marketing Spend", y="Revenue", data=data, height=5, col="Promo", ci=False,
               line_kws={"color":"red"},
               scatter_kws={"color":"#4EEEF9"})
```

```python
# Example 4: Relationship between Marketing Spend and Revenue - by column - by day(col_wrap) - add jitter too (x_jitter)
ax = sns.lmplot(x="Marketing Spend", y="Revenue", data=data, height=5, col="Day_Name", ci=False,
               line_kws={"color":"red"},
               scatter_kws={"color":"black"},
               col_wrap=3)
```
