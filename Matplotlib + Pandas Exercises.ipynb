```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from collections import counter
```

```python
url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv"
titanic = pd.read_csv(url, sep=",")
titanic.head()
```

```python
#Q. Set PassengerID as the Index.
titanic = titanic.set_index("PassengerId")
titanic.head()
```

```python
#Q. Create a pie chart presenting the male/female position.

#find the number of male to female first.
male = (titanic["Sex"] == "male").sum()
female = (titanic["Sex"] == "female").sum()

ratio = [male, female]

#create the pie chart. 
plt.pie(ratio,labels=(["male","female"]),explode=(0.15,0), shadow=False, startangle=90, autopct="%1.1f%%")

# View the plot drop above
plt.axis("equal")

# Set Labels
plt.title("Sex Proportion")

# View the plot
plt.tight_layout()
plt.show
```

```python
#Q. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender.
import seaborn as sns

#create the scatterplot
lm = sns.lmplot(x="Age", y="Fare", data=titanic, hue="Sex", fit_reg=False)

# Set the title and labels
lm.set(title="Fare x Age")

axes = lm.axes
axes[0,0].set_ylim(-5,)
axes[0,0].set_xlim(-5,85)
```

```python
#Q. Create a histogram with the Fare payed.
#sort the values from the top to the least value and slice the first 5 items
Farepayed = titanic.Fare.sort_values(ascending=False)

# Create bins interval using numpy
binsVal = np.arange(0,600,10)
binsVal

# Create the plot
plt.hist(Farepayed, bins=binsVal)

# Set the title and labels
plt.xlabel("Fare")
plt.ylabel("Frequency")
plt.title("Fare Payed Histrogram")

# Show the plot
plt.show()
```

```python
url3 = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/09_Time_Series/Apple_Stock/appl_1980_2014.csv"
apple = pd.read_csv(url3, sep=",")
apple.head()
```

```python
#Q. Check out the type of the columns.
#apple.dtypes
```

```python
#Q. Transfer the Date Column as a Datetime type. 
apple["Date"] = pd.to_datetime(apple["Date"])
apple.dtypes
```

```python
#Q. Set the date as the index. 
apple = apple.set_index("Date")
apple
```

```python
#Q. Is there any duplicate dates?
apple.duplicated()
```

```python
#Q. Make the first entry the oldest date. 
apple.sort_values(by="Date", ascending=True)
```

```python
#Q. Get the last business day of each month. 
apple_month = apple.resample("BM").mean()
apple_month

#Resampling is necessary when you're given a dataset recorded in some time 
#interbal and you want to change the time interval to something else. The 
#syntax of resample is fairly straightforward:
#<DataFrame or Series>.resample(arguments).<aggregate function>. 
#Arguments can be found in https://towardsdatascience.com/using-the-pandas-resample-function-a231144194c4
```

```python
#Q. What is the difference between the first day and the oldest. 
(apple.index.max() - apple.index.min()).days
```

```python
#Q. Plot the "Adj Close" value. Set the size of the figure to 13.5 x 9 inches.
apple_open = apple["Adj Close"].plot(title= "Apple Stock")

fig = apple_open.get_figure()
fig.set_size_inches(13.5,9)
```

```python
url5 = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data"
wine = pd.read_csv(url5)
wine.head()
```

```python
del wine["1"]
```

```python
wine.head()
```

```python
#Q. Delete the first, fourth, seventh, nineth, eleventh, thirteenth and fourteenth.
del wine["2.43"]
del wine["2.8"]
del wine[".28"]
del wine["5.64"]
del wine["3.92"]
del wine["1065"]
```

```python
#Q. Assign names to the column. 
wine.columns = ["alcohol","malic_acid","alcalinity_of_ash","magnesium","flavaniods","proanthocyanins","hue"]
wine.head()
```

```python
#Q. Set the values of the first 3 rows from alcohol as NaN.
wine.iloc[:3,0:1] = np.nan
wine.head(3)
```

```python
#Q. Set the value of the rows 3 and 4 of magnesium as NaN.
wine.iloc[3:5,3:4] = np.nan
wine.head(6)
```

```python
#Q. Fill the value of NaN with the number 10 in alcohol and 100 in magnesium. 
wine["alcohol"] = wine["alcohol"].fillna(10)
wine["magnesium"] = wine["magnesium"].fillna(100)
```

```python
wine.head(6)
```

```python
#Q. Count the number of missing values. 
wine.isna().sum()
```

```python
#Q. Create an array of 10 random numbers up until 10. 
rand_num = np.random.randint(0, high=10, size=10)
rand_num
```

```python
#Q. Use the random numbers you generated as an index and assign NaN value to
#each of cell. 
wine.alcohol[rand_num] = np.nan
wine.head(10)
```

```python
#Q. How many missing values do we have?
wine.isnull().sum()
```

```python
#Q. Delete the rows that contain missing values. 
wine = wine.dropna(axis=0, how="any")
wine.head()
```

```python
#Q. Print only the non-null values in alcohol
mask = wine.alcohol.notnull()
mask
```

```python
wine.alcohol[mask]
```

```python
#Q. Reset the index, so it starts with 0 again.
wine = wine.reset_index(drop=True)
wine.head()
```
