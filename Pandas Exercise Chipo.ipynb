```python
url5 = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv"
crime = pd.read_csv(url5, sep=",")
crime.head()
```

```python
#what is the type of the columns?
crime.info()
```

```python
#convert the type of the column year to datetime64.
crime.Year = pd.to_datetime(crime.Year, format="%Y")
crime.info()
```

```python
#Set the Year column as the index of the dataframe. 
crime = crime.set_index("Year", drop=True)
```

```python
#Delete the Total column.
del crime["Total"]
crime.head()
```

```python
#Group the year by decades and sum the values

#use resample to sum each decade
crimes = crime.resample("10AS").sum()

#use resample to get the mac value only for the "Population" column
population = crime["Population"].resample("10AS").max()

#updating the "Population" column
crimes["Population"] = population

crimes
```

```python
import datetime

url9 = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data"
data9 = pd.read_csv(url9, parse_dates=[["Yr","Mo","Dy"]], delim_whitespace=True)
data9
```

```python
#Q. Year 2061? Do we really have the data from this year? Create a function to fix it and apply it. 
def fix_century(x):
    year = x.year -100 if x.year > 1989 else x.year
    return datetime.date(year, x.month, x.day)

data9["Yr_Mo_Dy"] = data9["Yr_Mo_Dy"].apply(fix_century)
data9.head()
```

```python
#Q. Set the right dates as the index. Pay attention at the data type, it should be datetime64[ns].
data9["Yr_Mo_Dy"] = pd.to_datetime(data9["Yr_Mo_Dy"])

data9 = data9.set_index("Yr_Mo_Dy")
data9.head()
```

```python
#Q. Compute how many values are missing from each location over the entire record. 
data9.isnull().sum()
```

```python
#Q.Compute how many non-missing values there are in total. 
data9.notnull().sum()
```

```python
#Q. Calculate the mean windspeeds of the windspeeds over all the locations and all the times. 
data9.sum().sum()/data9.notna().sum().sum()
```
